# 4. vòng lặp
- `while`
- `for` + `range(start,stop, step)`

```python
# while (vòng lặp vô hạn -> lặp đến khi điều kiện sai)
i = 0
while i < 7:
    if i == 6:
        continue # toi lan lap tiep theo
    if i == 3:
        pass # bo qua
    if i == 5:
        break # dung vong lap
    print(i)
    i += 1
```

```python
# range(stop) -> stop - 1
for i in range(10):
    print(i)
```

```python
# range(start, stop) -> stop - 1
for i in range(2, 10):
    print(i)
```

```python
# range(start, stop, step) -> stop - 1, step là bước nhảy (tính từ start)
for i in range(2, 10, 2): # chỉ lấy các giá trị chẵn (start: 2 + 2 lần bước nhảy)
    print(i)
```

# 5. danh sách(list)
- thêm/xóa/sửa phần tử trong danh sách

```python
# khai bao
students = ["Phong", "Hieu", "Thong", "Duy", "Nam", "Kiet"]
print(students)
```

```python
# do dai danh sach
print(len(students))
```

```python
# truy cap phan tu
students[2]
```

```python
# thêm phần tử
students.append("Diep")
students
```

```python
# chinh sửa phần tử
students[0] = "Bạn mới"
students
```

```python
# xóa phần tử
del students[0]
students

# remove(bao loi neu khong co gia tri)
students.remove("Bạn mới")
```

```python
# chèn phần tử mới
# insert(vị trí, giá trị mới)
students.insert(2, "Bạn rất mới")
students
```

# 6. string(xâu chuỗi)
- gộp chuỗi,sửa chuỗi, cắt chuỗi


```python
# gia tri string được đặt vào ""
firstName = "Diep"
lastName = "Âu"
```

```python
# gop chuoi (+)
fullName = firstName + lastName + "van tue van tue van tue"
fullName
```

```python
# sua chuoi (replace(giá trị cần sửa, giá trị đổi mới, [số lượng cần đổi]) => hàm có trả về nên cần BIẾN LƯU GIÁ TRỊ MỚI)
fullName = fullName.replace("van tue", "muon nam") # mặc định đổi 1 lần 
print(fullName)
```

```python
# Split(giá trị phân cách các phần tử) -> trả về danh sách
fullNameList = fullName.split(" ")
fullNameList
```

```python
# Duyệt phần tử
for char in fullName:
    print(char)
```

# 7. Hàm (function)
- có trả về (`return`)
- có tham số truyền vào (`parameter`)


```python
# khai báo
def sumInList(your_list): # hàm truyền vào 1 tham số: your_list
    sum = 0 
    for item in your_list:
        sum += item
    return sum # hàm trả về giá trị của biến sum (1 giá trị cụ thể)
```

```python
# gọi hàm
print(sumInList([12,0,0,1]))
```
